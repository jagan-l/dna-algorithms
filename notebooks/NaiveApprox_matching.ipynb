```python
def naive_approx_matching(p,t,maxHammingDistance=1):
    occurences = []
    for i in range(len(t) - len(p) + 1):
        nmm = 0
        for j in range(len(p)):
            if t[i+j] != p[j]:
                nmm += 1 #If mismatch, increment counter.
                if nmm > maxHammingDistance: #The moment we exceed the allowed mismatch budget, we stop checking this alignment
                    break
        if nmm <= maxHammingDistance:
            occurences.append((i,nmm)) #return pair where first element is the ffset of the match and second is the Hamming distance
    return occurences
```

```python
naive_approx_matching('needle','needle noodle niidle', maxHammingDistance=2)
```
